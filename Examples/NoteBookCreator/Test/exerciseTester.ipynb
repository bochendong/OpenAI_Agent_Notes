## Test 7: 社会学练习题


```python
response = await Runner.run(exercise_agent, "生成3-5道关于社会化的练习题，难度由浅入深，包含选择题、填空题和简答题")
```

```python
print(f"生成了 {len(response.final_output.exercises)} 道题目\n")
for i, exercise in enumerate(response.final_output.exercises, 1):
    print(f"=" * 80)
    print(f"第 {i} 题 - {type(exercise).__name__}")
    print(f"=" * 80)
    print(f"题目: {exercise.question}")
    if hasattr(exercise, 'options'):
        print(f"选项: {exercise.options}")
        print(f"正确答案: {exercise.correct_answer}")
    if hasattr(exercise, 'answer'):
        print(f"答案: {exercise.answer}")
    if hasattr(exercise, 'blanks'):
        print(f"答案: {exercise.blanks}")
    if hasattr(exercise, 'explanation') and exercise.explanation:
        print(f"解释: {exercise.explanation}")
    print()
```
